Mi az a string?

A string (szöveg) a Pythonban karakterek sorozata, amit idézőjelek közé írunk.
Lehet '...', "...", vagy több soros is '''...''' vagy """...""".

```python
s1 = "Hello"
s2 = 'Világ'
s3 = """Ez
egy
többsoros
szöveg"""
print(s1, s2)

# Megjegyzés: A Pythonban a stringek immutable-ek, azaz nem módosíthatók közvetlenül.
```

Indexelés – karakterek elérése

A string minden karaktere sorszámmal (indexszel) rendelkezik, 0-tól kezdve.

```python
s = "Python"
print(s[0])   # első karakter -> P
print(s[1])   # második karakter -> y
print(s[-1])  # utolsó karakter -> n

# Negatív index esetén a számolás a végéről indul
```

Szeletelés (slicing)

A szeletelés segítségével részleteket tudunk kivágni a szövegből.

```python
s = "programozás"
print(s[0:7])   # 0-tól 6-ig -> 'program'
print(s[:7])    # ugyanez
print(s[3:])    # 3-tól a végéig -> 'gramozás'
print(s[::-1])  # megfordítja a stringet -> 'sázomgargorp'

# Szintaxis: s[start:stop:lépés]
```

Összefűzés és ismétlés

Stringeket össze lehet fűzni (+), vagy ismételni (*).

```python
a = "Hello"
b = "World"
print(a + " " + b)  # Hello World
print(a * 3)        # HelloHelloHello
```

Bejárás ciklussal

A string iterálható, tehát végig lehet menni minden karakterén

```python
s = "Python"
for ch in s:
    print(ch, end="-")
# Hasznos, ha például megszámolunk valamit (pl. hány ’a’ van a szövegben).
```

Hossz meghatározása

A len() függvénnyel lekérhetjük a karakterek szám

```python
s = "almafa"
print(len(s))  # 6
```

Stringek összehasonlítása

A stringek betűrend szerint hasonlíthatók össze (<, >, ==).

```python
print("alma" == "alma")  # True
print("alma" < "banán")  # True, mert az 'a' előbb van, mint 'b'
```

Hasznos string metódusok

| Metódus                         | Jelentés / Mit csinál                                       | Példa                                             | Eredmény / Megjegyzés  |
| :------------------------------ | :---------------------------------------------------------- | :------------------------------------------------ | :--------------------- |
| **`len(s)`**                    | Visszaadja a string hosszát (karakterek számát)             | `len("alma")`                                     | `4`                    |
| **`s.lower()`**                 | Kisbetűssé alakítja a stringet                              | `"PythOn".lower()`                                | `"python"`             |
| **`s.upper()`**                 | Nagybetűssé alakítja a stringet                             | `"python".upper()`                                | `"PYTHON"`             |
| **`s.capitalize()`**            | Csak az első betű lesz nagy, a többi kicsi                  | `"heLLo".capitalize()`                            | `"Hello"`              |
| **`s.title()`**                 | Minden szó első betűje nagybetűs                            | `"python programozás".title()`                    | `"Python Programozás"` |
| **`s.strip()`**                 | Eltávolítja az elején és végén lévő szóközöket              | `"   alma   ".strip()`                            | `"alma"`               |
| **`s.lstrip()` / `s.rstrip()`** | Csak bal / csak jobb oldalról vág szóközt                   | `"   alma".lstrip()`                              | `"alma"`               |
| **`s.replace(keres, csere)`**   | Lecseréli a részszöveget                                    | `"almafa".replace("a","o")`                       | `"olmof"`              |
| **`s.find(részlet)`**           | Megadja az első előfordulás indexét, vagy `-1`-et, ha nincs | `"program".find("gra")`                           | `3`                    |
| **`s.count(részlet)`**          | Megszámolja, hányszor szerepel a részlet                    | `"banana".count("a")`                             | `3`                    |
| **`s.split(elválasztó)`**       | Feldarabolja a stringet listává                             | `"alma,körte".split(",")`                         | `['alma', 'körte']`    |
| **`elválasztó.join(lista)`**    | Összefűzi a lista elemeit stringgé                          | `",".join(["a","b","c"])`                         | `"a,b,c"`              |
| **`s.startswith(részlet)`**     | True, ha a string ezzel kezdődik                            | `"Python".startswith("Py")`                       | `True`                 |
| **`s.endswith(részlet)`**       | True, ha a string ezzel végződik                            | `"Python3".endswith("3")`                         | `True`                 |
| **`s.isalpha()`**               | Csak betűk esetén True                                      | `"alma".isalpha()`                                | `True`                 |
| **`s.isdigit()`**               | Csak számjegyek esetén True                                 | `"123".isdigit()`                                 | `True`                 |
| **`s.isalnum()`**               | Csak betűk + számjegyek esetén True                         | `"Python3".isalnum()`                             | `True`                 |
| **`s.isspace()`**               | Csak szóközökből / tabból / sortörésből áll                 | `"   ".isspace()`                                 | `True`                 |
| **`s.swapcase()`**              | Kis ↔ nagy betűk felcserélése                               | `"PyTHon".swapcase()`                             | `"pYthON"`             |
| **`s.center(szélesség)`**       | Középre igazítja a stringet adott szélességben              | `"alma".center(10)`                               | `'   alma   '`         |
| **`s.ljust(n)` / `s.rjust(n)`** | Balra / jobbra igazítja a stringet                          | `"alma".rjust(8)`                                 | `'    alma'`           |
| **`s.zfill(n)`**                | Nullákkal tölti fel a stringet balról a megadott hosszra    | `"42".zfill(5)`                                   | `"00042"`              |
| **`"r" in s`**                  | Ellenőrzi, benne van-e részlet                              | `"program".find("pro")` vagy `"pro" in "program"` | `True`                 |
| **`not in`**                    | Ellenőrzi, hogy nincs-e benne                               | `"x" not in "program"`                            | `True`                 |



String formázás

A stringbe változókat is beilleszthetünk.
Modern, rövid megoldás az f-string:

```python
nev = "Bence"
kor = 18
print(f"{nev} {kor} éves.")

# Régi, de még hasznos formák:

print("{} {} éves.".format(nev, kor))
print("%s %d éves." % (nev, kor))
```

Escape karakterek és speciális jelek

Ha egy stringben pl. idézőjelet, sortörést vagy tabulátort szeretnél használni, az ún. escape karaktereket kell alkalmazni (\ jellel kezdődnek

```python
print("Ez egy idézet: \"Hello!\"")
print("Első sor\nMásodik sor")
print("Oszlop1\tOszlop2")
```

Fontos escape karakterek:

| Jel  | Jelentés           |
| ---- | ------------------ |
| `\n` | sortörés           |
| `\t` | tabulátor          |
| `\\` | visszaperjel       |
| `\"` | idézőjel (dupla)   |
| `\'` | idézőjel (szimpla) |


Karakterkódolás és Unicode

```python
ch = "Á"
print(ord(ch))   # 193 -> Unicode kódpont
print(chr(193))  # Á

'''
ord() → karakter → szám
chr() → szám → karakter

Hasznos pl. kódolási, rejtjelezési feladatoknál (Caesar-kód, stb.)
'''
```

Stringek módosítása „új” string létrehozásával

Mivel a string nem módosítható (“immutable”), ezért ha meg akarsz változtatni benne valamit, új stringet kell létrehozni.

```python
s = "pithon"
s = s[:2] + "y" + s[3:]  # ‘python’
print(s)
```

isalpha(), isdigit(), isalnum(), isspace(), startswith(), endswith()

Ezek ellenőrző metódusok, érettségin nagyon gyakoriak.

```python
s = "Python3"
print(s.isalpha())   # False (szám is van benne)
print(s.isdigit())   # False
print(s.isalnum())   # True (betű+szám)
print("   ".isspace()) # True

print(s.startswith("Py"))  # True
print(s.endswith("3"))     # True
```

Formázott kiírás többféle módon

```python
pi = 3.14159
print(f"A pi értéke: {pi:.2f}")  # 2 tizedes
nev = "Bence"
print(f"{nev:>10}")  # jobbra igazítás
print(f"{nev:<10}")  # balra igazítás
```

(Haladó, de érdekes) – szövegdarabolás egyedi karakterek szerint

Ha pl. vesszővel elválasztott adatsorod van (CSV-szerű), ez jó példa:

```python
s = "alma,körte,barack"
gyumolcsok = s.split(",")
for gy in gyumolcsok:
    print(gy)
```
