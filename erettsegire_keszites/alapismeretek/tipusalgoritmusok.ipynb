Mi az a típusalgoritmus?

A típusalgoritmusok olyan általános megoldási minták, amelyek gyakran ismétlődő feladatokra adnak megoldást.
Ezek segítségével logikusan és hatékonyan lehet feldolgozni adatsorokat (pl. listákat, számokat, szövegeket).


A legfontosabb típusalgoritmusok

Összegzés

Megszámlálás

Eldöntés

Kiválasztás

Keresés

Maximum-kiválasztás

Összegzés algoritmus

Feladat: egy sorozat (pl. lista) elemeinek összegét kiszámítani.

Elv:

induló érték: összeg = 0

végigmegyünk a listán, minden elemet hozzáadunk

az eredményt a ciklus végén kiírjuk

```python
# Összegzés algoritmus
szamok = [3, 7, 2, 5, 10]

osszeg = 0
for szam in szamok:
    osszeg += szam

print("A számok összege:", osszeg)

# Típusfeladat: „Add meg a lista elemeinek összegét!” vagy „Számítsd ki a jegyek átlagát!”
```

Megszámlálás algoritmus

Feladat: megszámolni, hány elem felel meg egy feltételnek.

Elv:

induló érték: db = 0

ha egy elem teljesíti a feltételt, növeljük a számlálót

```python
# Megszámlálás algoritmus
szamok = [3, 7, 2, 5, 10]

db = 0
for szam in szamok:
    if szam > 5:
        db += 1

print("5-nél nagyobb számok száma:", db)

# Típusfeladat: „Hány páros szám van a listában?” vagy „Hány tanuló ért el 60% feletti eredményt?”
```

Eldöntés algoritmus

Feladat: meg kell állapítani, hogy van-e a listában olyan elem, amely megfelel egy feltételnek.

Elv:

induló érték: van = False

ha találunk megfelelőt → van = True és akár megszakíthatjuk a ciklust

```python
# Eldöntés algoritmus
szamok = [3, 7, 2, 5, 10]

van = False
for szam in szamok:
    if szam % 2 == 0:
        van = True
        break

if van:
    print("Van páros szám a listában.")
else:
    print("Nincs páros szám a listában.")
```

Típusfeladat: „Van-e 0 a listában?” vagy „Szerepel-e a 100-as érték?”

Kiválasztás algoritmus

Feladat: tudjuk, hogy létezik olyan elem, amely megfelel a feltételnek – meg kell keresni hol van (indexe vagy értéke).

Elv:

addig megyünk, amíg meg nem találjuk a keresett elemet

```python
# Kiválasztás algoritmus
szamok = [3, 7, 2, 5, 10]

i = 0
while szamok[i] % 2 != 0:  # amíg nem páros
    i += 1

print("Első páros szám:", szamok[i])
print("Indexe:", i)

# Típusfeladat: „Add meg az első 0 indexét!” vagy „Melyik a legelső páros szám?”
```

Keresés algoritmus

Feladat: meg kell keresni, hol található egy adott elem a listában.
Eltér a „kiválasztás”-tól, mert nem biztos, hogy létezik.

Elv:

ciklusban haladunk

ha megtaláltuk: index = i és kilépünk

ha nem találjuk: index = -1

```python
# Keresés algoritmus
szamok = [3, 7, 2, 5, 10]
keresett = 5

index = -1
for i in range(len(szamok)):
    if szamok[i] == keresett:
        index = i
        break

if index != -1:
    print(f"A {keresett} érték indexe: {index}")
else:
    print(f"A {keresett} érték nem található.")

# Típusfeladat: „Keressük meg egy név helyét a névsorban!”
```

Maximum-kiválasztás algoritmus

Feladat: a legnagyobb (vagy legkisebb) elem és annak helye meghatározása.

Elv:

induló érték: az első elem

ha találunk nagyobbat, elmentjük

a végén a maximum érték és index az eredmény

```python
# Maximum-kiválasztás algoritmus
szamok = [3, 7, 2, 5, 10]

max_ertek = szamok[0]
max_index = 0

for i in range(1, len(szamok)):
    if szamok[i] > max_ertek:
        max_ertek = szamok[i]
        max_index = i

print("A legnagyobb elem:", max_ertek)
print("Indexe:", max_index)

# Típusfeladat: „Melyik tanuló kapta a legjobb jegyet?” vagy „Melyik városban volt a legmagasabb hőmérséklet?”

# Minimum-kiválasztás algoritmus

szamok = [3, 7, 2, 5, 10]

min_ertek = szamok[0]
for szam in szamok:
    if szam < min_ertek:
        min_ertek = szam

print("A legkisebb szám:", min_ertek)
```

Összegzés

| Algoritmus              | Kérdés                          | Eredmény       |
| ----------------------- | ------------------------------- | -------------- |
| **Összegzés**           | Mennyi az elemek összege?       | Összeg         |
| **Megszámlálás**        | Hány elem felel meg?            | Darabszám      |
| **Eldöntés**            | Van-e megfelelő elem?           | Igen / Nem     |
| **Kiválasztás**         | Hol van az első megfelelő elem? | Index / Érték  |
| **Keresés**             | Hol van a keresett érték?       | Index vagy -1  |
| **Maximum-kiválasztás** | Melyik a legnagyobb?            | Érték és index |


Mi az a rekurzió?

A rekurzió azt jelenti, hogy egy függvény önmagát hívja meg a megoldás során.
Minden rekurzív algoritmusban két dolog szükséges:

Alapeset – amikor már nem hívjuk újra (megállás feltétele)

Rekurzív lépés – amikor a függvény önmagát hívja kisebb problémára

Egyszerű példa – visszaszámlálás

```python
def visszaszamlal(n):
    if n == 0:
        print("Start!")
    else:
        print(n)
        visszaszamlal(n - 1)

visszaszamlal(5)
```

Faktoriális rekurzívan

```python
def faktorialis(n):
    if n == 0:
        return 1
    else:
        return n * faktorialis(n - 1)

print(faktorialis(5))
```
