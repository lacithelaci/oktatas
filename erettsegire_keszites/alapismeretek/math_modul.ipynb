Mi a math modul?

A math modul a Python beépített matematikai függvényeit tartalmazza.
Segítségével elérhetünk olyan műveleteket, mint négyzetgyök, logaritmus, trigonometrikus függvények, stb.

A modult importálni kell, mielőtt használni kezdjük

```python
import math
```

Fontosabb konstansok
| Név        | Jelentés                  | Példa                                  |
| ---------- | ------------------------- | -------------------------------------- |
| `math.pi`  | π (pi) ≈ 3.14159          | `math.pi * r ** 2` – kör területe      |
| `math.e`   | Euler-féle szám ≈ 2.71828 | `math.e ** x` – exponenciális számítás |
| `math.inf` | Végtelen érték            | `math.inf > 10**100` → `True`          |
| `math.nan` | „Not a Number” (nem szám) | lebegőpontos hibák esetén használható  |


```python
import math

print("math.pi =", math.pi)
print("math.e =", math.e)
print("math.inf =", math.inf)
print("math.nan =", math.nan)
print("math.isinf(math.inf) =", math.isinf(math.inf))
print("math.isnan(math.nan) =", math.isnan(math.nan))
```

Kerekítési és abszolút érték függvények
| Függvény        | Leírás                            | Példa              | Kimenet |
| --------------- | --------------------------------- | ------------------ | ------- |
| `math.ceil(x)`  | Felfelé kerekít                   | `math.ceil(3.2)`   | `4`     |
|                 |                                   | `math.ceil(-1.8)`  | `-1`    |
| `math.floor(x)` | Lefelé kerekít                    | `math.floor(3.8)`  | `3`     |
|                 |                                   | `math.floor(-2.3)` | `-3`    |
| `math.trunc(x)` | Egész rész levágása (nem kerekít) | `math.trunc(3.9)`  | `3`     |
|                 |                                   | `math.trunc(-3.9)` | `-3`    |
| `math.fabs(x)`  | Abszolút érték (float típussal)   | `math.fabs(-5)`    | `5.0`   |


```python
import math

print("math.ceil(3.2) =", math.ceil(3.2))
print("math.ceil(-1.8) =", math.ceil(-1.8))
print("math.floor(3.8) =", math.floor(3.8))
print("math.floor(-2.3) =", math.floor(-2.3))
print("math.trunc(3.9) =", math.trunc(3.9))
print("math.trunc(-3.9) =", math.trunc(-3.9))
print("math.fabs(-5) =", math.fabs(-5))
```

Hatványozás, gyökvonás, logaritmus
| Függvény            | Leírás                      | Példa               | Kimenet             |
| ------------------- | --------------------------- | ------------------- | ------------------- |
| `math.pow(x, y)`    | (x^y) – hatványozás         | `math.pow(2, 3)`    | `8.0`               |
| `math.sqrt(x)`      | Négyzetgyök                 | `math.sqrt(25)`     | `5.0`               |
| `math.exp(x)`       | (e^x)                       | `math.exp(1)`       | `2.718281828459045` |
| `math.log(x)`       | Természetes logaritmus (ln) | `math.log(math.e)`  | `1.0`               |
| `math.log(x, base)` | Logaritmus adott alappal    | `math.log(100, 10)` | `2.0`               |
| `math.log10(x)`     | 10-es alapú logaritmus      | `math.log10(1000)`  | `3.0`               |
| `math.log2(x)`      | 2-es alapú logaritmus       | `math.log2(8)`      | `3.0`               |


```python
import math

print("math.pow(2, 3) =", math.pow(2, 3))
print("math.sqrt(25) =", math.sqrt(25))
print("math.exp(1) =", math.exp(1))
print("math.log(math.e) =", math.log(math.e))
print("math.log(100, 10) =", math.log(100, 10))
print("math.log10(1000) =", math.log10(1000))
print("math.log2(8) =", math.log2(8))
```

Trigonometrikus függvények
| Függvény       | Leírás           | Példa                   | Kimenet                    |
| -------------- | ---------------- | ----------------------- | -------------------------- |
| `math.sin(x)`  | szinusz          | `math.sin(math.pi / 2)` | `1.0`                      |
| `math.cos(x)`  | koszinusz        | `math.cos(0)`           | `1.0`                      |
| `math.tan(x)`  | tangens          | `math.tan(math.pi / 4)` | `1.0`                      |
| `math.asin(x)` | inverz szinusz   | `math.asin(1)`          | `1.5707963267948966 (π/2)` |
| `math.acos(x)` | inverz koszinusz | `math.acos(0)`          | `1.5707963267948966 (π/2)` |
| `math.atan(x)` | inverz tangens   | `math.atan(1)`          | `0.7853981633974483 (π/4)` |

Fok ↔ Radián átváltás
| Függvény          | Példa                   | Kimenet             |
| ----------------- | ----------------------- | ------------------- |
| `math.degrees(x)` | `math.degrees(math.pi)` | `180.0`             |
| `math.radians(x)` | `math.radians(180)`     | `3.141592653589793` |


```python
import math

print("math.sin(math.pi/2) =", math.sin(math.pi/2))
print("math.cos(0) =", math.cos(0))
print("math.tan(math.pi/4) =", math.tan(math.pi/4))
print("math.asin(1) =", math.asin(1))
print("math.acos(0) =", math.acos(0))
print("math.atan(1) =", math.atan(1))
print("math.degrees(math.pi) =", math.degrees(math.pi))
print("math.radians(180) =", math.radians(180))
```

Kombinatorikai és faktoriális függvények
| Függvény            | Leírás                      | Példa               | Kimenet |
| ------------------- | --------------------------- | ------------------- | ------- |
| `math.factorial(x)` | Faktoriális (x!)            | `math.factorial(5)` | `120`   |
| `math.comb(n, k)`   | Kombinációk száma (C(n, k)) | `math.comb(5, 2)`   | `10`    |
| `math.perm(n, k)`   | Permutációk száma (P(n, k)) | `math.perm(5, 2)`   | `20`    |


```python
import math

print("math.factorial(5) =", math.factorial(5))
print("math.comb(5, 2) =", math.comb(5, 2))
print("math.perm(5, 2) =", math.perm(5, 2))
```

Egyéb hasznos függvények

| Függvény               | Leírás                                   | Példa                       | Kimenet |
| ---------------------- | ---------------------------------------- | --------------------------- | ------- |
| `math.isfinite(x)`     | Véges szám-e                             | `math.isfinite(5)`          | `True`  |
| `math.isinf(x)`        | Végtelen-e                               | `math.isinf(math.inf)`      | `True`  |
| `math.isnan(x)`        | NaN érték-e                              | `math.isnan(math.nan)`      | `True`  |
| `math.copysign(x, y)`  | `x` abszolút értékét `y` előjelével adja | `math.copysign(3, -1)`      | `-3.0`  |
| `math.hypot(x, y)`     | Pitagoraszi távolság: √(x²+y²)           | `math.hypot(3, 4)`          | `5.0`   |
| `math.dist(p, q)`      | Távolság két pont között                 | `math.dist((1, 2), (4, 6))` | `5.0`   |
| `math.remainder(x, y)` | Maradék (IEEE módon)                     | `math.remainder(7, 3)`      | `1.0`   |


```python
import math

print("math.isfinite(5) =", math.isfinite(5))
print("math.isinf(math.inf) =", math.isinf(math.inf))
print("math.isnan(math.nan) =", math.isnan(math.nan))
print("math.copysign(3, -1) =", math.copysign(3, -1))
print("math.hypot(3, 4) =", math.hypot(3, 4))
print("math.dist((1, 2), (4, 6)) =", math.dist((1, 2), (4, 6)))
print("math.remainder(7, 3) =", math.remainder(7, 3))
```
