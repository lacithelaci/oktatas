Értékadás

```python
a = 5 # Értéket az = jellel adunk a változónak, A bal oldalon mindig a változó neve, a jobb oldalon az érték vagy kifejezés áll.
b = 3
osszeg = a + b

# Értékadó (assignment) operátorok példák

x = 10
print("Kezdő érték:", x)

x += 5      # x = x + 5
print("x += 5 →", x)

x -= 3      # x = x - 3
print("x -= 3 →", x)

x *= 2      # x = x * 2
print("x *= 2 →", x)

x /= 4      # x = x / 4
print("x /= 4 →", x)

x //= 2     # x = x // 2
print("x //= 2 →", x)

x %= 3      # x = x % 3
print("x %= 3 →", x)

x **= 2     # x = x ** 2
print("x **= 2 →", x)

```

Aritmetikai operátorok
| Művelet          | Jel  | Példa    | Eredmény |
| ---------------- | ---- | -------- | -------- |
| Összeadás        | `+`  | `5 + 3`  | `8`      |
| Kivonás          | `-`  | `5 - 3`  | `2`      |
| Szorzás          | `*`  | `5 * 3`  | `15`     |
| Osztás (valós)   | `/`  | `5 / 2`  | `2.5`    |
| Egész osztás     | `//` | `5 // 2` | `2`      |
| Maradékos osztás | `%`  | `5 % 2`  | `1`      |
| Hatványozás      | `**` | `2 ** 3` | `8`      |


```python
# Aritmetikai operátorok példák

a = 10
b = 3

print("a =", a)
print("b =", b)
print()

print("Összeadás: a + b =", a + b)
print("Kivonás: a - b =", a - b)
print("Szorzás: a * b =", a * b)
print("Osztás (valós): a / b =", a / b)
print("Egész osztás: a // b =", a // b)
print("Maradékos osztás: a % b =", a % b)
print("Hatványozás: a ** b =", a ** b)

# Extra: kifejezések
print("\nKifejezés: (a + b) * 2 =", (a + b) * 2)
print("Kifejezés: a / (b + 2) =", a / (b + 2))
```

Összehasonlító (relációs) operátorok
| Művelet              | Jel  | Példa    | Eredmény |
| -------------------- | ---- | -------- | -------- |
| Egyenlő              | `==` | `5 == 3` | `False`  |
| Nem egyenlő          | `!=` | `5 != 3` | `True`   |
| Kisebb               | `<`  | `5 < 3`  | `False`  |
| Nagyobb              | `>`  | `5 > 3`  | `True`   |
| Kisebb vagy egyenlő  | `<=` | `5 <= 3` | `False`  |
| Nagyobb vagy egyenlő | `>=` | `5 >= 3` | `True`   |


```python
# Relációs (összehasonlító) operátorok példák

a = 5
b = 8

print("a =", a)
print("b =", b)
print()

print("Egyenlő (==):", a == b)
print("Nem egyenlő (!=):", a != b)
print("Kisebb (<):", a < b)
print("Nagyobb (>):", a > b)
print("Kisebb vagy egyenlő (<=):", a <= b)
print("Nagyobb vagy egyenlő (>=):", a >= b)

# Extra példa
print("\nFelhasználói példa:")
print("Van-e kedvezmény, ha életkor >= 18?")
eletkor = 16
print("Életkor:", eletkor)
print("Kedvezmény:", eletkor >= 18)
```

Logikai operátorok
| Operátor | Jelentés | Példa            | Eredmény |
| -------- | -------- | ---------------- | -------- |
| `and`    | és       | `True and False` | `False`  |
| `or`     | vagy     | `True or False`  | `True`   |
| `not`    | negálás  | `not True`       | `False`  |


```python
# Logikai operátorok példák

a = True
b = False

print("a =", a)
print("b =", b)
print()

print("AND (és): a and b =", a and b)
print("OR (vagy): a or b =", a or b)
print("NOT (negálás): not a =", not a)
print("NOT b =", not b)

# Összetett logikai kifejezés
x = 12
print("\nx =", x)
print("x nagyobb 10-nél és páros?", x > 10 and x % 2 == 0)
print("x kisebb 10-nél vagy páros?", x < 10 or x % 2 == 0)
print("x nem páros?", not (x % 2 == 0))
```

```python
# Egyéb gyakori operátorok: tagság, azonosítás, string konkatenáció

# Tagsági operátorok (in, not in)
nev = "Anna"
print("'A' in nev:", 'A' in nev)
print("'x' not in nev:", 'x' not in nev)

# Azonosítás operátorok (is, is not)
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print("\nA listák tartalma egyforma?", a == c)
print("De ugyanaz az objektum?", a is c)
print("És b is ugyanarra mutat?", a is b)

# Szövegösszefűzés (concatenation)
elso = "Hello"
masodik = "Világ"
print("\nSzövegösszefűzés:", elso + " " + masodik)

# Ismétlés (* operátor)
print("Ismétlés:", "Hi! " * 3)
```

| `int()`   | szövegből egész        | `int("5")`                              | `5`      |
| `float()` | szövegből valós        | `float("3.14")`                         | `3.14`   |

```python
# KOMMENTEK (megjegyzések) példák

# Egy soros komment – a # jellel kezdődik
print("Ez egy program")  # Ez a sor végén lévő komment

# Többsoros megjegyzés több # jellel:
# Ez egy hosszabb leírás
# A program célja: egy egyszerű kiírás gyakorlása
print("Python programozás alapok")

# Alternatíva (nem igazi komment, de dokumentációként használható):
"""
Ez egy többsoros szöveg (docstring),
amit a Python nem kommentként kezel, de nem hajt végre,
így gyakran használják leírásként függvények elején.
"""

def osszeg(a, b):
    """Ez a függvény két számot ad össze és visszaadja az eredményt."""
    return a + b

print(osszeg(3, 4))
```

```python
# KIFEJEZÉSEK KIÉRTÉKELÉSI SORRENDJE (precedencia, asszociativitás)

# Példa alap:
eredmeny = 5 + 3 * 2
print("5 + 3 * 2 =", eredmeny)  # Először a szorzás: 3*2 = 6, majd 5 + 6 = 11

# Zárójelezés megváltoztatja a sorrendet:
eredmeny = (5 + 3) * 2
print("(5 + 3) * 2 =", eredmeny)  # Először az összeadás: 8*2 = 16

# Több művelet sorrendben:
eredmeny = 2 + 3 * 4 ** 2
print("2 + 3 * 4 ** 2 =", eredmeny)
# Itt: 4 ** 2 = 16 → 3 * 16 = 48 → 2 + 48 = 50

# Balról jobbra haladás (asszociativitás):
eredmeny = 100 / 10 * 2
print("100 / 10 * 2 =", eredmeny)  # (100 / 10) * 2 = 20.0 → nem 100 / (10*2)

# Logikai műveletek sorrendje
x = True
y = False
z = True

eredmeny = x or y and z
print("True or False and True =", eredmeny)
# Értelmezés: 'and' erősebb → False and True = False → True or False = True

# Egyenlőség- és relációs vizsgálatok:
a = 3
b = 5
c = 5

print("a < b == c :", a < b == c)
# Értelmezés: Python láncolja: (a < b) és (b == c) → True and True → True
```
