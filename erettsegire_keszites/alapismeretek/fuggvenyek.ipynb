Függvények alapfogalma

A függvény egy logikailag összefüggő utasításblokk, amit többször is fel tudunk használni.

Két fő típusa:

Eljárás: nem ad vissza értéket (return nélkül)

Függvény: visszatér egy értékkel (return használata)

Függvény definiálása
def → a függvény definíció kulcsszava

koszontes → függvény neve

() → paraméterlista, itt üres

A függvény nem ad vissza értéket, csak végrehajtja a print-et

```python
def koszontes():
    print("Szia!")

koszontes() # Függvény (eljárás) meghívása
```

Paraméterek

A paraméterek értékeket adnak át a függvénynek, amit a függvény fel tud használni.

```python
def koszontes_nevvel(nev):
    print("Szia,", nev)

koszontes_nevvel("Anna")
koszontes_nevvel("Béla")

# Paraméter alapértelmezett értékkel is rendelkezhet
def koszontes_nevvel(nev="Vendég"):
    print("Szia,", nev)

koszontes_nevvel()      # Vendég
koszontes_nevvel("Anna")  # Anna

# Több paramétere is lehet egy függvénynek
def osszeg(a, b):
    print(a + b)

osszeg(3, 5)  # 8
```

Visszatérési értékek (return)

A return kulcsszóval a függvény értéket ad vissza, amit elmenthetünk változóba vagy tovább használhatunk.

```python
def osszeg(a, b):
    return a + b

eredmeny = osszeg(4, 7)
print("Eredmény:", eredmeny)
```

Beépített függvények
| Függvény  | Mit csinál?       | Példa                       |
| --------- | ----------------- | --------------------------- |
| `print()` | Kiír a képernyőre | `print("Hello")`            |
| `len()`   | Hossz             | `len([1,2,3])` → 3          |
| `type()`  | Adat típusa       | `type(5)` → `<class 'int'>` |
| `max()`   | Legnagyobb érték  | `max(4,9,2)` → 9            |
| `min()`   | Legkisebb érték   | `min(4,9,2)` → 2            |
| `sum()`   | Összeg            | `sum([1,2,3])` → 6          |
| `round()` | Kerekítés         | `round(3.14159,2)` → 3.14   |


Type hintek (típus jelzés)

Python dinamikus típusú, de a type hintek segítik az olvashatóságot és a hibakeresést.

Nem kötelező, de érdemes használni

a: int, b: int → a paraméter típusa

-> int → a visszatérési érték típusa

```python
def osszeg(a: int, b: int) -> int:
    return a + b

print(osszeg(3, 5))
```
