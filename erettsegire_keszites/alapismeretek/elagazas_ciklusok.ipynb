Elágazás

Az if szerkezet lehetővé teszi, hogy a program feltétel alapján válasszon a különböző utasítások közül.

A feltétel logikai értékre (True/False) értékelődik.

```python
kor = 17

if kor >= 18:
    print("Felnőtt")
else:
    print("Kiskorú")
```

```python
szam = 0

if szam > 0:
    print("Pozitív")
elif szam == 0: # elif = „egyébként ha” - Több elif is lehet, de az első igaz ág lefutása után minden további ág elmarad.
    print("Nulla")
else:
    print("Negatív")
```

```python
print("Felnőtt") if kor >= 18 else print("Kiskorú") # Egy soros if
```

For ciklus

Iterációs ciklus, amely lista, string, tuple vagy range elemein halad végig. Mindig meghatározott lépésszámon keresztül fut

```python
lista = [1, 2, 3]
for elem in lista:
    print(elem, end=" ")
else:
    print()
# vagy
for elem in range(len(lista)):
    print(lista[elem], end= " ")

```

```python
# Beágyazott for ciklus
for i in range(2):
    for j in range(3):
        print(i, j)
```

Break – a ciklus megszakítása

A break azonnal kilép a ciklusból, függetlenül attól, hogy hány iteráció van hátra.

Használható for és while ciklusban is.

Ha van else ág a ciklusnál, break esetén az else nem fut le.

```python
for i in range(1, 6):
    if i == 4:
        break
    print("i =", i)
else:
    print("Ciklus véget ért")
```

Continue – a jelenlegi iteráció kihagyása

A continue átugorja a ciklus aktuális lépését, és a következő iterációra ugrik.

A ciklus nem szakad meg, csak a jelenlegi lépés nem fut le teljesen.

```python
for i in range(1, 6):
    if i % 2 == 0:
        continue
    print("i (páratlan):", i)
```

While ciklus

A while addig fut, amíg a feltétel True.

Feltétel lehet tetszőleges logikai kifejezés.

```python
szam = 1
while szam <= 3:
    print(szam)
    szam += 1
```
