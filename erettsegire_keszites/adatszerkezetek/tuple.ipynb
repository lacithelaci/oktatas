## 1. Mi az a tuple?

A tuple egy **rendezett, változtathatatlan (immutable) adattípus**, amely több elemet tud tárolni.  

- Rendezett → van indexelés (0-tól kezdődően)  
- Változtathatatlan → egyszer létrehozva **nem módosítható**  

```python
# Tuple létrehozása
my_tuple = (1, 2, 3, 4)
print(my_tuple)  # Output: (1, 2, 3, 4)
```

## 2. Tuple vs List

| Tulajdonság       | List               | Tuple               |
|------------------|------------------|-------------------|
| Módosíthatóság    | Igen              | Nem               |
| Szintaxis         | `[1, 2, 3]`       | `(1, 2, 3)`       |
| Használat         | Dinamikus adattár | Rögzített adatok  |
| Teljesítmény      | Lassabb           | Gyorsabb          |

## 4. Elemszám és indexelés

```python
t = (10, 20, 30, 40)

# Indexelés
print(t[0])  # 10
print(t[-1]) # 40

# Szeletelés
print(t[1:3])  # (20, 30)
```

## 5. Tuple műveletek

```python
# összeadás
t1 = (1, 2)
t2 = (3, 4)
t3 = t1 + t2
print(t3)  # (1, 2, 3, 4)

# Ismétlés

t = (0, 1) * 3
print(t)  # (0, 1, 0, 1, 0, 1)

# Elem keresése / index / count
t = (1, 2, 3, 2)
print(2 in t)        # True
print(t.index(3))    # 2
print(t.count(2))    # 2
```

## Tuple kibontás

```python
# Minden tuple ugyanakkora hosszúságú (3 elem)
tuples_list = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]

# Bejárás
for t in tuples_list:
    # Tuple unpacking
    a, b, c = t
    print(f"a: {a}, b: {b}, c: {c}")
```
