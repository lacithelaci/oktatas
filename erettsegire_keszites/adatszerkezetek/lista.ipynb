# Python List (Lista) Bemutató

---

## 1. Mi az a lista?

A lista egy **rendezett, módosítható adattípus**, amely tetszőleges számú elemet tud tárolni.  
- Rendezett → van indexelés (0-tól kezdődően)  
- Módosítható → elemek hozzáadhatók, módosíthatók, törölhetők  

```python
# Lista létrehozása
fruits = ["apple", "banana", "cherry"]
print(fruits)
```

## 2. Elemek elérése

```python
fruits = ["apple", "banana", "cherry"]

# Indexelés
print(fruits[0])   # apple
print(fruits[-1])  # cherry

# Szeletelés
print(fruits[0:2])  # ['apple', 'banana']
```

## 3. Elem hozzáadása és módosítása

```python
fruits = ["apple", "banana"]

# Új elem hozzáadása a végére
fruits.append("cherry")

# Több elem hozzáadása
fruits.extend(["orange", "kiwi"])

# Elem beszúrása adott pozícióra
fruits.insert(1, "mango")

# Elem módosítása
fruits[0] = "pear"

print(fruits)
```

## 4. Elem törlése

```python
fruits = ["apple", "banana", "cherry"]

# Index alapján
del fruits[1]

# Visszaadott érték alapján
fruit = fruits.pop(0)
print(fruit)

# Érték alapján
fruits.remove("cherry")

# Lista teljes törlése
fruits.clear()
```

## 5. Lista bejárása

```python
fruits = ["apple", "banana", "cherry"]

# Egyszerű bejárás
for fruit in fruits:
    print(fruit)

# Index és érték
for i, fruit in enumerate(fruits):
    print(i, fruit)
```

## 6. Lista műveletek

```python
a = [1, 2]
b = [3, 4]

# Összefűzés
c = a + b

# Ismétlés
d = a * 3

# Ellenőrzés
print(2 in a)  # True

# Hossz
print(len(a))  # 2
```

## 7. Nested list (Beágyazott lista)

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Bejárás
for row in matrix:
    for value in row:
        print(value, end=" ")
    print()
```
