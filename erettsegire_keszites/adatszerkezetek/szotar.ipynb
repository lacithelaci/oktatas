## 1. Mi az a dictionary?

A dictionary egy **rendezett (Python 3.7+), kulcs-érték párokat tároló adattípus**.  
- Kulcsok egyediek (unique) és változtathatatlanok (immutable) lehetnek: pl. string, int, tuple.  
- Értékek tetszőleges típusúak lehetnek.  

```python
# Dictionary létrehozása
person = {"name": "Alice", "age": 25, "city": "Budapest"}
print(person)
```

## 2. Elemek elérése

```python
person = {"name": "Alice", "age": 25, "city": "Budapest"}

# Kulcs alapján
print(person["name"])  # Alice

# Biztonságos elérés .get()-tel
print(person.get("country", "Ismeretlen"))  # Ismeretlen
```

## 3. Elem hozzáadása és módosítása

```python
person = {"name": "Alice", "age": 25}

# Új kulcs hozzáadása
person["city"] = "Budapest"

# Érték módosítása
person["age"] = 26

print(person)
```

## 4. Elem törlése

```python
person = {"name": "Alice", "age": 25, "city": "Budapest"}

# Kulcs törlése
del person["age"]

# Visszaadja az eltávolított értéket
city = person.pop("city")
print(city)  # Budapest

# Az összes elem törlése
person.clear()
print(person)  # {}
```

## 5. Dictionary bejárása

```python
person = {"name": "Alice", "age": 25, "city": "Budapest"}

# Kulcsok bejárása
for key in person:
    print(key, person[key])

# Értékek bejárása
for value in person.values():
    print(value)

# Kulcs-érték párok bejárása
for key, value in person.items():
    print(f"{key}: {value}")
```

## 6. Dictionary kulcsok, értékek, elemek

```python
person = {"name": "Alice", "age": 25, "city": "Budapest"}

# Kulcsok listája
keys = list(person.keys())
print(keys)

# Értékek listája
values = list(person.values())
print(values)

# Elempárok listája
items = list(person.items())
print(items)
```
