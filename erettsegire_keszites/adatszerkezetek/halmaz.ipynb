## 1. Mi az a halmaz?

A set egy **rendezés nélküli, egyedi elemeket tároló adattípus**.  
- Rendezés nincs → nincs indexelés  
- Minden elem egyedi  
- Használható a halmazelméleti műveletekre  

```python
# Halmaz létrehozása
fruits = {"apple", "banana", "cherry"}
print(fruits)
```

## 2. Elek hozzáadása és eltávolítása

```python
fruits = {"apple", "banana"}

# Elem hozzáadása
fruits.add("cherry")

# Több elem hozzáadása
fruits.update(["orange", "kiwi"])

# Elem eltávolítása
fruits.remove("banana")  # KeyError ha nincs
fruits.discard("banana") # Nem dob hibát

# Véletlenszerű elem eltávolítása
fruit = fruits.pop()
print(fruit)

# Halmaz törlése
fruits.clear()
```

## 3. Halmaz bejárása

```python
fruits = {"apple", "banana", "cherry"}

for fruit in fruits:
    print(fruit)
```

## 4. Halmaz műveletek

```python
a = {1, 2, 3}
b = {3, 4, 5}

# Egyesítés
print(a | b)  # {1,2,3,4,5}

# Metszet
print(a & b)  # {3}

# Különbség
print(a - b)  # {1,2}

# Szimmetrikus különbség
print(a ^ b)  # {1,2,4,5}
```
