# Cleaning Fifa-18 Dataset

Dataset source: https://www.kaggle.com/thec03u5/fifa-18-demo-player-dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('CompleteDataset.csv')
```

```python
pd.set_option('display.max_columns',None) # alapérteémezetten 20 oszlopot jelenít meg
pd.options.display.max_rows
```

```python
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Index'})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
subdf1 = df.loc[:,'Special':'LWB']
subdf2 = df.loc[:,'RAM':]
subdf = pd.concat([subdf1,subdf2], axis=1) # axis --> egymás után vagy mellé 0 - sor 1 - oszlop
subdf.head()
```

```python
subdf = subdf.astype('str').replace('\+.*','',regex=True).replace('\-.*','',regex=True)
```

```python
subdf = subdf.astype('float64')
```

```python
subdf.head()
```

```python
df[subdf.columns] = subdf
```

```python
df['Position'] = df['Preferred Positions'].str.split(' ').str[0]
```

```python
df.info()
```

```python
df.to_csv('fifa18-cleaned.csv',index=None)
```
