```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('stations.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['lat'] = df['Latitude'].map(lambda x: (x+100)%200 - 100) # origo x kiszámítása
df['long'] = df['Longitude'].map(lambda x: (x+200)%400 - 200) # origo y kiszámítása
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['long'], y=df['lat'],marker='+')
```

```python
X = df[['long','lat']]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X);
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['long'], y=df['lat'],marker='+', hue=kmeans.labels_, palette="Set1")
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=5,min_samples=50) # eps: adatpontok közti max távolság, min samples: a középpont körüli pontok minimális értéke
db.fit(X)
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['long'], y=df['lat'],marker='+', hue=db.labels_, palette="Set1")
```

```python
db = DBSCAN(eps=5,min_samples=30) # eps: adatpontok közti max távolság, min samples: a középpont körüli pontok minimális értéke
db.fit(X)
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['long'], y=df['lat'],marker='+', hue=db.labels_, palette="Set1")
```

```python
db = DBSCAN(eps=3,min_samples=50) # eps: adatpontok közti max távolság, min samples: a középpont körüli pontok minimális értéke
db.fit(X)
plt.figure(figsize=(20,10))
sns.scatterplot(x=df['long'], y=df['lat'],marker='+', hue=db.labels_, palette="Set1")
```
