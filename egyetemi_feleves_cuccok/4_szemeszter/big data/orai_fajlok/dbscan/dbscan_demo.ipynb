```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# CSV beolvasása
df = pd.read_csv('dbscan_demo_data.csv')
X_total = df[['x', 'y']].values

print(f"Adatok betöltve, összesen {X_total.shape[0]} pont")
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X_total[:,0], X_total[:,1], c='blue', s=50, alpha=0.6)
plt.title("DBSCAN demo adathalmaz")
plt.xlabel("X koordináta")
plt.ylabel("Y koordináta")
plt.show()
```

```python
# DBSCAN paraméterek: eps = 0.9, min_samples = 10
db = DBSCAN(eps=0.9, min_samples=10)
db.fit(X_total)

labels = db.labels_  # -1 jelöli a zajpontokat

# Összes klaszter száma (-1 a zaj)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Klaszterek száma: {n_clusters}")
print(f"Zajpontok száma: {n_noise}")
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X_total[:,0], X_total[:,1], c=labels, s=50, alpha=0.6)
plt.title("DBSCAN klaszterezés")
plt.xlabel("X koordináta")
plt.ylabel("Y koordináta")
plt.show()
```
