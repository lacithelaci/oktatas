# One-Hot Encoding for Features

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('vehicles.csv')
```

```python
df.head()
```

```python
df['drive'] = df['drive'].fillna('EMPTY')
```

Let's check the `drive` column unique values

```python
drive_values = df['drive'].unique()
drive_values
```

```python
len(drive_values)
```

```python
np.expand_dims(drive_values,-1).shape
drive_values
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(df[['drive']])
```

```python
encoder.categories_
```

```python
encoder.transform([['2-Wheel Drive']]).toarray()
```

```python
encoder.transform([['4-Wheel Drive']]).toarray()
```

```python
encoder.transform([['4-Wheel or All-Wheel Drive']]).toarray()
```
