### Pandas display opciók

A pandas `display` beállításai azt szabályozzák, hogyan jelennek meg a DataFrame-ek (az adatokat nem módosítják).

- **`display.max_columns`**  
  Meghatározza, hogy legfeljebb hány oszlop jelenjen meg.  
  `None` → minden oszlop  
  konkrét szám (pl. `20`) → max. ennyi oszlop

- **`display.max_rows`**  
  Meghatározza, hogy legfeljebb hány sor jelenjen meg.  
  `None` → minden sor  
  konkrét szám (pl. `100`) → max. ennyi sor

Az aktuális értékek a `pd.options.display` alatt érhetők el.

```python
pd.set_option('display.max_columns',None)
pd.options.display.max_rows
```

```python
df = df.rename(columns={'Unnamed: 0': 'Index'}) # oszlopok átnevezése
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
subdf1 = df.loc[:,'Special':'LWB']
subdf2 = df.loc[:,'RAM':]
subdf = pd.concat([subdf1,subdf2], axis=1) # axis=1 esetén egymás mellé, 0 esetén egymás alá teszi az oszlopokat
subdf.head()
```

```python
subdf = subdf.astype(str).replace(r'\+.*', '', regex=True).replace(r'\-.*', '', regex=True) # regex, levágja +/- jeleket
```

```python
subdf = subdf.astype('float64')
```

```python
subdf.head()
```

```python
df[subdf.columns] = subdf
```

```python
df['Position'] = df['Preferred Positions'].str.split(' ').str[0]
```

```python
df.info()
```

```python
df.to_csv('fifa18-cleaned.csv',index=None,sep=',')
```

```python
not_gk = df.copy()
not_gk.dropna(subset=['CB','CM'], how='all') # csak azokat törli, ahol mindkettő üres,
                                             #'any paraméter esetén viszont vagy feltétel van közöttük'
```

```python
gk = df[df['Position']=="GK"].copy() # Ha részhalmazon dolgoztok, mindig készítsetek másolatot

gk['CAM'] = gk['CAM'].fillna(0)
gk['Acceleration'] = gk['Acceleration'].fillna(df['Acceleration'].mean())
gk.head()
```
