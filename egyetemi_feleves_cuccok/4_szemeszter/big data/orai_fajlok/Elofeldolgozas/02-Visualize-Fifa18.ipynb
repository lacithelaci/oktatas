### Using FacetGrid

**FacetGrid (Seaborn):**  
Lehetővé teszi, hogy egy adatvizualizációt **többszörösen szeleteljünk kategóriák szerint**, és minden szelethez külön grafikont készítsünk.  

**Fő paraméterek:**
- `data` → a vizualizálandó DataFrame  
- `row` → a sorokhoz tartozó kategória (minden egyes érték külön sort kap)  
- `col` → az oszlopokhoz tartozó kategória (minden egyes érték külön oszlopot kap)  
- `hue` → színezéshez, csoportok elkülönítéséhez  
- `height` → egy subplot magassága  
- `aspect` → szélesség/magasság arány, így a grafikon alakja beállítható  

**Használat:**  
Miután létrehoztad a FacetGrid-et, a `.map()` függvénnyel megadható, **melyik plot típust** és **melyik változót** ábrázolja minden szeletben.

```python
g = sns.FacetGrid(df, row='Position', height=2, aspect=4)
g.map(sns.kdeplot, "Age")
```

### Age vs Rank

**lineplot paraméterek:**
- `estimator` → meghatározza, hogyan aggregáljuk az y értékeket az x értékekhez (pl. `mean`, `median`).
- `errorbar` → konfidencia intervallum a vonal körül; `None` kikapcsolja.

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Age',y='Overall',estimator='mean',errorbar=('ci', 95))
```

```python
data = df[df['Position'].isin(['GK','ST','CF'])]

plt.figure(figsize=(12,5))
sns.lineplot(data=data, x='Age',y='Overall',estimator='mean',errorbar=None, hue='Position')
```

### Top 50 Teams

```python
fig, ax = plt.subplots(figsize=(12,4))
top50clubs = df[['Club','Overall']].groupby('Club').mean().sort_values(by='Overall', ascending=False)[:50]
top50clubs.plot.bar(ax=ax)
```

```python
plt.figure(figsize=(12,4))
chart = sns.barplot(x=top50clubs.index.values, y=top50clubs['Overall'])
chart.set_xticklabels(chart.get_xticklabels(), rotation=90);
```

### Fix Values

```python
column = 'Value'

for i in range(df.shape[0]):
    value = str(df.loc[i,column]).replace('€','')
    if value.endswith('M'):
        value=float(value[:-1])
    elif value.endswith('K'):
        value=float(value[:-1]) / 1000
    df.loc[i,column] = value
```

```python
df['Value'] = df['Value'].astype('float64')
```

```python
df['Value'].dtype
```

```python
column = 'Wage'

for i in range(df.shape[0]):
    value = str(df.loc[i,column]).replace('€','')
    if value.endswith('K'):
        value=float(value[:-1])
    df.loc[i,column] = value
```

```python
df['Wage'] = df['Wage'].astype('float64')
```

```python
df['Wage'].dtype
```

```python
df.head()
```

### Plot Values

```python
fig, axes = plt.subplots(figsize=(20,10), nrows=2)
sns.boxplot(data=df, x='Value', ax=axes[0]); axes[0].set_xlabel('Value in €M')
sns.boxplot(data=df, x='Wage', ax=axes[1]); axes[1].set_xlabel('Wage in €K')
```
