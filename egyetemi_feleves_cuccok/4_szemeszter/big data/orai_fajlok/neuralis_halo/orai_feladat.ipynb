## SMS Spam Collection Dataset

Ez a dataset SMS üzeneteket tartalmaz, amelyeket spam (kéretlen) és ham (valódi) kategóriákba soroltak. Összesen 5574 angol nyelvű üzenetből áll, ahol minden sor egy üzenetet és annak címkéjét tartalmazza.

A célja a spam szűrés (text classification) modellek tanításának és tesztelésének támogatása.

## 1. feladat – Adatok beolvasása

Olvasd be a `spam_ham.csv` fájlt egy pandas DataFrame-be.  
A fájl vesszővel (`,`) van elválasztva, ezért ennek megfelelően add meg a beolvasásnál a szeparátort.

A beolvasás után jelenítsd meg az első néhány sort, hogy ellenőrizd az adatok sikeres betöltését.

## 2. feladat – Szövegvektorizálás és adatok előkészítése

Hozz létre egy `TfidfVectorizer` objektumot, amely a 3000 leggyakoribb kifejezést veszi figyelembe.  

Ezután:
- az `v2` oszlopot mentsd el az **X** változóba (ez lesz a feature),
- az `v1` oszlopot mentsd el az **Y** változóba (ez lesz a célváltozó),
- végül az X-en hajts végre TF-IDF transzformációt.

## 3. feladat – Adathalmaz felosztása

Oszd fel az adathalmazt tanuló (train) és teszt (test) halmazra.

A cél, hogy a modell tanítására és kiértékelésére külön adatokat használjunk, így elkerülhető a túltanulás és objektíven mérhető a teljesítmény.

Használj megfelelő arányt (pl. 80%-20%) és állíts be `random_state` értéket a reprodukálhatóság érdekében.

## 4. feladat – Naive Bayes modell tanítása és kiértékelése

Tanítsunk egy Naive Bayes (MultinomialNB) modellt a tanító adathalmazon, majd értékeljük ki a teljesítményét a teszt halmazon.

A kiértékeléshez használjunk pontosságot (accuracy) és részletes osztályozási riportot.
